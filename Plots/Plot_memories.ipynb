```python
import numpy as np
import matplotlib.pyplot as plt
# Generate random integers (0 or 1) and cast to float
test_memories = np.random.randint(0, 2, size=(5, 32)).astype(float)
test_partitioned_memories = []

beta = 0.5
for memory in test_memories:
    memory = np.exp(memory/beta)
    memory = memory / np.sum(memory)
    test_partitioned_memories.append(memory)
print("Partitioned memories:")
for idx, p_memory in enumerate(test_partitioned_memories):
    print(f"Memory {idx+1}: {p_memory}")

fig, ax = plt.subplots(figsize=(12, 2.5))

im = ax.imshow(
    test_partitioned_memories,
    cmap="Greys",
    aspect="auto",
    interpolation="none",
)

ax.set_yticks(np.arange(5))
ax.set_yticklabels(
    [f"Memory {i+1}" for i in range(5)],
    fontsize=14
)

ax.set_xticks(np.arange(0, 32, 4))
ax.set_xticklabels(np.arange(1, 33, 4), fontsize=12)

ax.set_xlabel("Component Index", fontsize=14)
ax.set_ylabel("Memory", fontsize=14)

# draw grid lines between components
ax.set_xticks(np.arange(-0.5, 32, 1), minor=True)
ax.set_yticks(np.arange(-0.5, 5, 1), minor=True)
ax.grid(which="minor", color="lightgray", linewidth=0.5)

for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```
